```python
#On télécharge le fichier cif 
with MPRester("2N9dqDh3wLHCSK1O") as m:
    struct = m.get_structure_by_material_id("mp-754541")
    
w = CifWriter(struct)
w.write_file('TmTlO2_mp-754541_symmetrized.cif')
```
