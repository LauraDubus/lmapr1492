```python
from pymatgen.ext.matproj import MPRester
from pymatgen.io.cif import CifWriter
from pymatgen.symmetry.analyzer import *
from pymatgen.io.cif import *
from pymatgen.core.operations import *
from pymatgen.symmetry.site_symmetries import *
import matplotlib.pyplot as plt
import matplotlib.image as img
```

```python
#On télécharge le fichier cif 
with MPRester("2N9dqDh3wLHCSK1O") as m:
    struct = m.get_structure_by_material_id("mp-754541")
    
w = CifWriter(struct)
w.write_file('TmTlO2_mp-754541_symmetrized.cif')
```

```python
#On définit les trois opérations de symmétries que l'on va analyser
Sym_1  = SymmOp.from_xyz_string('-y, -x, z') #symétrie 8
Sym_2  = SymmOp.from_xyz_string('-x, -y, -z') #symétrie 2
Sym_3  = SymmOp.from_xyz_string('y, -x+y, -z') #symétrie 4
print ("Il existe 12 opérations de symétries pour cette structure. Les trois opérations de symétries que nous allons étudier sont les suivantes, avec à chaque fois la matrice de rotation et la matrice de translation de chaque opération:")
print ("1)")
print (Sym_1)
print ("2)")
print (Sym_2)
print ("3)")
print (Sym_3)
```

```python
#Première opération de symétrie
coordonées_initiales=[0.33333,0.66667,0.16667] #atome 1
coordonées_finales = Sym_1.operate(coordonées_initiales) #sym8
print("La première opération de symétrie est la suivante, avec sa matrice de rotation et sa matrice de translation :")
print(Sym_1)
print("Les coordonées initiales de l'atome choisi sont :")
print (coordonées_initiales)
print("Les coordonées de l'atome après avoir effectué la première opération de symétrie sont :")
print(coordonées_finales)
print("Voici la représentation de l'opération de symétrie (qui est un plan miroir) sur base de jupyter-jsmol:")
image_1=img.imread('sym1.jpg')
fig, axes = plt.subplots(figsize=(10,10))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_1)
plt.show()
```

```python
#Seconde opération de symétrie
coordonées_initiales=[0.33333,0.66667,0.66667] #atome 6
coordonées_finales = Sym_2.operate(coordonées_initiales) #sym2
print("La seconde opération de symétrie est la suivante, avec sa matrice de rotation et sa matrice de translation :")
print(Sym_2)
print("Les coordonées initiales de l'atome choisi sont :")
print (coordonées_initiales)
print("Les coordonées de l'atome après avoir effectué la deuxième opération de symétrie sont :")
print(coordonées_finales)
print("Voici la représentation de l'opération de symétrie sur base de jupyter-jsmol:")
image_2=img.imread('sym2.10.png')
fig, axes = plt.subplots(figsize=(20,20))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_2)
plt.show()
```

```python
#Troisième opération de symétrie
coordonées_initiales=[0.66667,0.33333,0.11047] #atome 8
coordonées_finales = Sym_3.operate(coordonées_initiales) #sym4
print("La troisième opération de symétrie est la suivante, avec sa matrice de rotation et sa matrice de translation :")
print(Sym_3)
print("Les coordonées initiales de l'atome choisi sont :")
print (coordonées_initiales)
print("Les coordonées de l'atome après avoir effectué la troisième opération de symétrie sont :")
print(coordonées_finales)
print("Voici la représentation de l'opération de symétrie sur base de jupyter-jsmol:")
image_2=img.imread('sym3.2.png')
fig, axes = plt.subplots(figsize=(20,20))
axes.get_xaxis().set_visible(False)
axes.get_yaxis().set_visible(False)
axes.imshow(image_2)
plt.show()
```
